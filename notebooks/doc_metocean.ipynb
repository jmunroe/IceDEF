```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML

from importlib import reload
from icedef import metocean

%matplotlib inline
```

```python
start_date = np.datetime64('2015-04-23')
end_date = start_date + np.timedelta64(1, 'D')
date_bounds = start_date, end_date
```

# Create Metocean Objects

```python
atmosphere = metocean.Atmosphere(date_bounds, model='ECMWF')
```

```python
ocean = metocean.Ocean(date_bounds)
```

# Animate Velocity Fields

```python
wind_animation = atmosphere.animate_field()
```

```python
HTML(wind_animation.to_html5_video())
```

```python
ocean_animation = ocean.animate_field()
```

```python
HTML(ocean_animation.to_html5_video())
```
